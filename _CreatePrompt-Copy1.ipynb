```python
class Prompt():
    def __init__(self, task, files):        
        self.task = task
        self.files = files        

    def read_python_file(self, file_path: str, tag: str = "file") -> str:
        try:
            with open(file_path, 'r', encoding='utf-8') as f:
                content = f.read()
            return f"<{tag} name='{file_path}'>\n{content}\n</{tag}>"
        except Exception as e:
            return f"<error>{e}</error>"


    def create_prompt(self):
        output_string = f"""
        <task>
        {self.task}
        </task>
        """
        for file in self.files:
            output_string += self.read_python_file(file)

        return output_string          
```
